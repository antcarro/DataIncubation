```python
links_dict=dict()
years_of_transcripts=['https://www.cbsnews.com/news/face-the-nation-transcripts-'+str(i)+'/' for i in range(2014,2019)]
for directory in years_of_transcripts:
    page=requests.get(directory).text
    links=re.finditer(r'<a href="(https://www.cbsnews.com/news/[^"]*?transcript[^"]*?)"')
    links=re.finditer(r'(face-the-nation-transcripts-([a-z]*-[0-9]*-[0-9]*)-[a-z0-9\-]*)',page)
    for l in links:
        print(l.group(2))
        links_dict[l.group(2)]='https://www.cbsnews.com/news/'+l.group(1)
```

```python
def text_cleaner(text):
    #First, we isolate the dialogue, which always lies between the "<div data-page> tags
    #k=re.finditer(r'<div data-page.*?>(.*?)</div>',text)
    #T=''
    #for K in k:
    #    T+=K.group(1)
    #Step -1: remove newline characters
    T=re.sub(r'(\\n|\n)',' ',text)
    T=re.sub(r'(\\r|\r)',' ',T)
    #Step 0: short-hand face the nation
    T=re.sub(r'FACE THE NATION','ftn',T)
    #Step 2: Remove paragraph tags, newline characters, and \r s
    T=re.sub(r'<.*?>',' ',T)
    T=re.sub(r'(<|>)',' ',T)
    #Step 3: Remove video clips (since they have no interactions)
    T=re.sub(r'\(BEGIN .*?\).*?\(END .*?\)',' ',T)
    #Step 5: Remove publication titles \\"NAME\\"
    T=re.sub(r'\\"[^\\]*\\"','',T)
    #Step 6: Remove unicode
    #Step 6.5: Remove parentheticals (names used sometimes) and replace (CROSSTALK) with something
    T=re.sub(r'\(CROSSTALK\)',':crosstalk:',T)
    T=re.sub(r'\(.*?\)',' ',T)
    T=re.sub(r'\\u[0-9]{4}','',T)
    #Step 7: Place a > in front of each speaker
    T=re.sub(r'([A-Z][\.\s,A-Z]*:)','>'+r'\1',T)
    #Step 8: Remove , because they get in the way
    T=re.sub(r',','',T)
    #Step 9: Remove spaces between anything and :
    T=re.sub(r'\s+:',':',T)
    return T
def text_analyzer(transcript):
    D=dict()
    """returns a dictionary of speakers, and analyzes whether they were interrupted"""
    lines=re.finditer(r'>([^:]*):([^{^>]*)',transcript)
    last_interrupted=False #keeps track of whether previous speaker was interrupted by current speaker
    for line in lines:
        speaker=line.group(1).replace(' ','')#speaker
        monologue=line.group(2)#what they said
        num_words=len(monologue.split(' '))
        interrupted=bool(re.search(r'(\.{3}|-{2})\s*$',monologue))
        if speaker in D:
            D[speaker]['num_lines']+=1
            D[speaker]['num_words']+=num_words
            if last_interrupted:
                D[speaker]['num_interrupts']+=1
            D[speaker]['num_interrupted']+=interrupted
            
        else:
            D[speaker]={'num_lines':1,'num_words':num_words,'num_interrupted': int(interrupted),'num_interrupts':int(last_interrupted)}
        last_interrupted=interrupted
    return D 
```

```python
def large_corpus_cleaner(links_dict):
    host_dict=dict()
    for l in links_dict:
        l_page = requests.get(links_dict[l]).text
        T=text_cleaner(l_page)
        with open(l,'w+') as f:
            f.write(T)
        host_dict[l]=host_summary(text_analyzer(T))
    return host_dict
    #host,interruptions=host_and_interruptions(text_analyzer(text_cleaner(l_page)))
    #analysis_dict[l]=(host,interruptions)
```

```python
def host_summary(show_dict):
    host=''
    host_lines=0
    total_lines=0
    for a in show_dict:
        speaker_lines=show_dict[a]['num_lines']
        total_lines+=show_dict[a]['num_lines']
        if speaker_lines>host_lines:
            host_lines=show_dict[a]['num_lines']
            host=a    
    return (host,show_dict[host]['num_lines'],
            show_dict[host]['num_interrupted'],
            show_dict[host]['num_interrupts'])
```

```python
host_dict=large_corpus_cleaner(links_dict)
```

```python
df=pd.DataFrame.from_dict(host_dict,orient='index')
df.columns=['Host','Average Lines','Hosts Interrupted','Hosts Interrupting']
```

```python
def name_replacement(short_name,new_name):
    if short_name in new_name or new_name in short_name:
        return short_name
    else:
        return new_name
```

```python
for name in ['GARRETT','SCHIEFFER','ODONNELL','ROSE','DICKERSON']:
    df['Host']=df['Host'].apply(lambda x: name_replacement(name,x))
```
